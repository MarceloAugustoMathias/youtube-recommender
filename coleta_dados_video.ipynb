```python
import pandas as pd
import numpy as np

import youtube_dl

pd.set_option('display.max_columns', 73)
```

# Coleta de Dados

```python
ydl = youtube_dl.YoutubeDL({"ignoreerrors" : True})
```

Palavras-chave de busca no youtube

```python
querys = ['machine+learning', 'data+science', 'kaggle']
```

### Extração de dados do youtube com a biblioteca *youtube_dl*

```python
resultados = []

for query in querys:
    r = ydl.extract_info("ytsearchdate200:{}".format(query), download = False)
    for entry in r['entries']:
        if entry is not None:
            entry['query'] = query
    resultados += r['entries']
```

```python
resultados = [e for e in resultados if e is not None]
len(resultados)
```

# Criação do Dataframe

```python
df = pd.DataFrame(resultados)

df['upload_date'] = pd.to_datetime(df['upload_date'])
df['tempo_desde_pub'] = (pd.to_datetime("2021-09-01") - df['upload_date']) / np.timedelta64(1, 'D')

df.head()
```

### Exportação dos dados para tagueamento / labeling

```python
colunas = ['title', 'upload_date', 'view_count', 'tempo_desde_pub']
df[colunas].to_csv('raw_data_without_labels.csv')
```
